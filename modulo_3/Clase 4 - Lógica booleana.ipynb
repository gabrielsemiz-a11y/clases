## 1. True y False

En Python existe un tipo de dato especial llamado **booleano** (`bool`). Solo tiene dos valores posibles: `True` y `False`.

| Pregunta | Resultado posible |
|---|---|
| ¿Hay pan? | `True` / `False` |
| ¿Está lloviendo? | `True` / `False` |
| ¿El prompt está vacío? | `True` / `False` |
| ¿El modelo está cargado? | `True` / `False` |


````python
# Valores booleanos en Python

hay_pan = True
esta_lloviendo = False
modelo_cargado = True

print(hay_pan)
print(esta_lloviendo)
print(modelo_cargado)
print(type(modelo_cargado))

# Atención: Python distingue mayúsculas y minúsculas.
# True  -> correcto
# False -> correcto
# true  -> error
# false -> error
````

## 3. Operadores de comparación

Las comparaciones evalúan dos valores y devuelven `True` o `False`.

| Operador | Significado | Ejemplo |
|---|---|---|
| `==` | igual a | `tipo_tarea == "resumir"` |
| `!=` | distinto de | `prompt != ""` |
| `>` | mayor que | `temperatura > 30` |
| `<` | menor que | `minutos < 5` |
| `>=` | mayor o igual | `tokens_estimados >= 1000` |
| `<=` | menor o igual | `tokens_estimados <= limite_contexto` |

`=` asigna un valor. `==` compara dos valores.


````python
# Comparaciones con datos de los ejercicios de Clase 3

minutos = 4
humedad_suelo = 25
temperatura = 34
prompt = "Explicá qué es un modelo local en 3 oraciones"
tokens_estimados = 40
limite_contexto = 2048

print("¿Hay poco tiempo?", minutos < 5)
print("¿La humedad es baja?", humedad_suelo < 30)
print("¿Hace calor?", temperatura > 30)
print("¿El prompt tiene texto?", prompt != "")
print("¿El prompt entra en contexto?", tokens_estimados <= limite_contexto)
````

## 4. Operadores lógicos: `and`, `or`, `not`

Los operadores lógicos combinan condiciones.

| Operador | Cuándo devuelve `True` | Ejemplo |
|---|---|---|
| `and` | Cuando todas las condiciones son verdaderas | `hay_pan and quiere_algo_caliente` |
| `or` | Cuando al menos una condición es verdadera | `color == "rojo" or color == "verde"` |
| `not` | Invierte el valor | `not esta_lloviendo` |

Tablas de verdad:

| A | B | A and B |             | A | B | A or B |
|---|---|---|---|---|---|---|
| True | True | True |             | True | True | True |
| True | False | False |             | True | False | True |
| False | True | False |             | False | True | True |
| False | False | False |             | False | False | False |      



````python
# Operadores lógicos aplicados a los ejercicios

hay_pan = True
quiere_algo_caliente = True
hay_fruta = False

print("Veamos las respuestas:")
print("- ¿Puedo preparar tostadas? =", hay_pan and quiere_algo_caliente)
print("- ¿Tengo alguna opción rápida? =", hay_pan or hay_fruta)
print("- ¿No hay fruta? =", not hay_fruta)

color = "verde"
peso = 180
textura = "lisa"

es_manzana = peso > 150 and (color == "rojo" or color == "verde")
es_naranja = color == "naranja" and textura == "rugosa"

print()
print("Reglas para determinar si es manzana o naranja:")
print(f"Color: {color}, Peso: {peso}, Textura: {textura}")
print()
print("¿Es manzana según la regla?", es_manzana)
print("¿Es naranja según la regla?", es_naranja)
````

## 5. Paréntesis y precedencia

Python evalúa los operadores lógicos en este orden:

1. `not`
2. `and`
3. `or`

Cuando una expresión tiene más de dos condiciones, conviene usar paréntesis para que la intención sea clara.

Estas dos expresiones pueden dar resultados distintos:

```python
pesado and rojo or verde
pesado and (rojo or verde)
```

La segunda expresa mejor la regla: la fruta debe ser pesada y además roja o verde.


````python
# Paréntesis: la intención importa

pesado = False
rojo = False
verde = True

resultado_sin_parentesis = pesado and rojo or verde
resultado_con_parentesis = pesado and (rojo or verde)

print("Sin paréntesis:", resultado_sin_parentesis)
print("Con paréntesis:", resultado_con_parentesis)

print("Lectura sin paréntesis: (pesado and rojo) or verde")
print("Lectura con paréntesis: pesado and (rojo or verde)")
````

| Incluir historial | `necesita_contexto_anterior and historial_disponible` |
| Revisar respuesta | `tema_sensible or confianza_baja` |

Estas condiciones todavía no ejecutan acciones. Solo responden `True` o `False`. En Clase 5 se convertirán en ramas con `if`.
